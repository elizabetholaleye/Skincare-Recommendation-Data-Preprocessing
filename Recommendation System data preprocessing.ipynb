COLLABORATIVE FILTERING

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load your dataset from a specific path
file_path =  r'C:\Users\owner\Desktop\400 Level\FINAL YEAR PROJECT\Data Preprocessing\cleanedcosmetics2.csv' # Replace with your actual file path
df = pd.read_csv(file_path)

# Define functions for collaborative filtering 

def get_cf_recommendations(user_id, user_item_matrix, item_similarity, top_n=3):
    user_ratings = user_item_matrix.loc[user_id].values.reshape(1, -1)
    similarity_scores = np.dot(user_ratings, item_similarity).flatten()

    # Get indices of top-N similar items
    similar_indices = similarity_scores.argsort()[::-1]
    
    # Exclude items already rated by the user
    rated_items = user_item_matrix.loc[user_id].to_numpy().nonzero()[0]
    similar_indices = [idx for idx in similar_indices if idx not in rated_items]

    # Get top-N recommendations
    number_of_recommendations = 3
    recommended_items = list(user_item_matrix.columns[similar_indices][:number_of_recommendations])

    return recommended_items

# Create user-item interaction matrix
user_item_matrix = df.pivot_table(index='user_id', columns='Name', values='Rank', fill_value=0)

# Compute item similarity matrix
item_similarity = cosine_similarity(user_item_matrix.T)

# Example user ID for whom we want recommendations
user_id = 15

# Get recommendations using collaborative filtering
recommendations = get_cf_recommendations(user_id, user_item_matrix, item_similarity)
print(f"Recommendations for User {user_id} (CF): {recommendations}")
```

```python
# Hybrid Filtering
def hybrid_recommendations(user_id, product, user_item_matrix, item_similarity, top_n=5):
    content_recs = content_recommendations(product, top_n=top_n)
    cf_recs = get_cf_recommendations(user_id, user_item_matrix, item_similarity, top_n=top_n)
    
    # Combine recommendations and remove duplicates
    #hybrid_recs = list(dict.fromkeys(content_recs + cf_recs))

    # Ensure both content_recs and cf_recs are lists of strings
    content_recs = [str(item) for item in content_recs]
    cf_recs = [str(item) for item in cf_recs]
    
    # Combine recommendations and remove duplicates
    hybrid_recs = content_recs + cf_recs
    hybrid_recs = list(dict.fromkeys(hybrid_recs))  # Remove duplicates
    
    
    return hybrid_recs[:top_n]

# Example usage
user_id = 20
product = 'The Rice Polish Foaming Enzyme Powder'  # Replace with an actual product name from your dataset

# Get hybrid recommendations
recommendations = hybrid_recommendations(user_id, product, user_item_matrix, item_similarity)
print(f"Hybrid Recommendations for User {user_id} based on '{product}': {recommendations}")
```
